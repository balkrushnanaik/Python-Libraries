# Advanced Pandas: 

## What is Pandas?
- Pandas is a popular open-source Python library used for data analysis and manipulation. It provides powerful tools for working with structured data such as tables, spreadsheets, and CSV files.
## The two main pandas data structures are:
1. Series – a one-dimensional labeled array.
2. DataFrame – a two-dimensional table with rows and columns (similar to a spreadsheet).

```python
# import pandas
import pandas as pd
pd.__version__
```

## 1. Create DataFrame: 
- A DataFrame is the most important data structure in pandas. It is a 2-dimensional table with rows and columns, similar to an Excel spreadsheet or a SQL table.

```python
data = {
    'Name': ['Madhav', 'Vishakha', 'Lalita', 'Rishabh'],
    'Age': [16,17,18,19],
    'Salary': [90000, 70000, 50000, 30000]
}
df = pd.DataFrame(data)
print(df)
```
