```python
# Create a DataFrame for Pivot Table Practice


import pandas as pd

data = {
    "Order_ID": [101,102,103,104,105,106,107,108,109,110,
                 111,112,113,114,115,116,117,118,119,120],

    "Customer": ["Amit","Priya","Rahul","Sneha","Rohit",
                 "Anjali","Karan","Neha","Vikas","Pooja",
                 "Amit","Priya","Rahul","Sneha","Rohit",
                 "Anjali","Karan","Neha","Vikas","Pooja"],

    "Gender": ["Male","Female","Male","Female","Male",
               "Female","Male","Female","Male","Female",
               "Male","Female","Male","Female","Male",
               "Female","Male","Female","Male","Female"],

    "City": ["Pune","Mumbai","Delhi","Pune","Mumbai",
             "Delhi","Pune","Mumbai","Delhi","Pune",
             "Mumbai","Delhi","Pune","Mumbai","Delhi",
             "Pune","Mumbai","Delhi","Pune","Mumbai"],

    "Category": ["Electronics","Clothing","Furniture","Electronics","Clothing",
                 "Furniture","Electronics","Clothing","Furniture","Electronics",
                 "Clothing","Furniture","Electronics","Clothing","Furniture",
                 "Electronics","Clothing","Furniture","Electronics","Clothing"],

    "Product": ["Laptop","Shirt","Table","Mobile","Jeans",
                "Chair","Headphones","T-Shirt","Sofa","Tablet",
                "Jacket","Bed","Camera","Shoes","Cupboard",
                "Smart Watch","Kurta","Desk","Speaker","Hoodie"],

    "Quantity": [2,5,1,3,4,2,6,5,1,2,
                 3,1,2,4,2,3,5,2,4,6],

    "Price": [55000,1200,8000,25000,1800,
              3500,3000,900,15000,22000,
              2500,18000,45000,3000,20000,
              12000,1500,7000,5000,2200],

    "Sales": [110000,6000,8000,75000,7200,
              7000,18000,4500,15000,44000,
              7500,18000,90000,12000,40000,
              36000,7500,14000,20000,13200]
}

df = pd.DataFrame(data)
df.head()
```

1. Create a pivot table showing total **Sales** by **City**.


```python
pivot = pd.pivot_table(
    df,
    values="Sales",
    index="City",
    aggfunc="sum"
)

print(pivot)
```

```python
# 2. Find total **Quantity** sold by **Category**.
pivot = pd.pivot_table(df,index='Category',values='Quantity',aggfunc='sum')
pivot
```
