In Pandas, **string methods** are used to perform operations on text data in a Series. They are accessed using the `.str` accessor.

## Sample Data

```python
import pandas as pd

df = pd.DataFrame({
    'Name': ['Balkrushna', 'John Doe', 'alice', 'PYTHON']
})

print(df)
```

Output:

```
         Name
0  Balkrushna
1    John Doe
2       alice
3      PYTHON
```

---

# Common String Methods in Pandas

## 1. `str.lower()`

Converts text to lowercase.

```python
df['Name'].str.lower()
```

Output:

In Pandas, **wildcards are usually used with regular expressions (regex)** in string operations such as `str.contains()`, `str.extract()`, `str.replace()`, and `str.match()`.

## Sample Data

```python
import pandas as pd

s = pd.Series([
    'apple',
    'banana',
    'grape',
    'pineapple',
    'orange'
])
```

---

# Common Wildcards (Regex Patterns)

| Wildcard | Meaning                      | Example   |
| -------- | ---------------------------- | --------- |
| `.`      | Any single character         | `a.e`     |
| `*`      | Zero or more occurrences     | `ap*`     |
| `+`      | One or more occurrences      | `ap+`     |
| `?`      | Zero or one occurrence       | `colou?r` |
| `^`      | Starts with                  | `^a`      |
| `$`      | Ends with                    | `e$`      |
| `.*`     | Any number of any characters | `a.*e`    |
| `[abc]`  | Match a, b, or c             | `[ae]`    |
| `[^abc]` | Not a, b, or c               | `[^a]`    |
| `\d`     | Any digit                    | `\d+`     |
| `\w`     | Letter, digit, underscore    | `\w+`     |

---

## 1. Find Strings Starting with "a"

```python
s[s.str.contains(r'^a')]
```

Output:

```python
0    apple
dtype: object
```

---

## 2. Find Strings Ending with "e"

```python
s[s.str.contains(r'e$')]
```

Output:

```python
0        apple
2        grape
3    pineapple
4       orange
dtype: object
```

---

## 3. Any Characters Between "a" and "e"

Pattern:

```python
r'a.*e'
```

```python
s[s.str.contains(r'a.*e')]
```

Output:

```python
0        apple
2        grape
3    pineapple
4       orange
```

Explanation:

* `a` → starts matching from `a`
* `.*` → any characters
* `e` → ends matching at `e`

---

## 4. Match Exactly One Character

```python
s[s.str.contains(r'a.p')]
```

Examples matched:

```text
app
alp
aXp
```

`.` = any single character.

---

## 5. Strings Containing "an"

```python
s[s.str.contains(r'.*an.*')]
```

Output:

```python
1    banana
4    orange
```

---

## 6. Find Digits Using Wildcards

```python
data = pd.Series(['A12', 'B45', 'XYZ'])
```

```python
data.str.contains(r'\d')
```

Output:

```python
0     True
1     True
2    False
```

---

## 7. Extract Email Domains

```python
emails = pd.Series([
    'abc@gmail.com',
    'xyz@yahoo.com'
])

emails.str.extract(r'@(.+)')
```

Output:

```python
           0
0  gmail.com
1  yahoo.com
```

## SQL Wildcards vs Pandas Regex

| SQL     | Pandas Regex |
| ------- | ------------ |
| `%abc%` | `.*abc.*`    |
| `abc%`  | `^abc`       |
| `%abc`  | `abc$`       |
| `_abc`  | `.abc`       |

Example:

SQL:

```sql
WHERE name LIKE '%apple%'
```

Pandas:

```python
df[df['name'].str.contains('apple')]
```

# Pandas `concat()` - All Operations in One Program

```python
import pandas as pd

print("=" * 60)
print("PANDAS CONCAT() OPERATIONS")
print("=" * 60)

# -------------------------------------
# Create DataFrames
# -------------------------------------
df1 = pd.DataFrame({
    "ID": [101, 102],
    "Name": ["Alice", "Bob"],
    "Age": [25, 30]
})

df2 = pd.DataFrame({
    "ID": [103, 104],
    "Name": ["Charlie", "David"],
    "Age": [35, 40]
})

df3 = pd.DataFrame({
    "ID": [105, 106],
    "Salary": [50000, 60000]
})

print("\nDataFrame 1")
print(df1)

print("\nDataFrame 2")
print(df2)

print("\nDataFrame 3")
print(df3)

# ==================================================
# 1. Vertical Concatenation (Default - axis=0)
# ==================================================
print("\n1. Vertical Concatenation")
vertical = pd.concat([df1, df2])
print(vertical)

# ==================================================
# 2. Vertical Concatenation with New Index
# ==================================================
print("\n2. Vertical Concatenation (ignore_index=True)")
vertical_index = pd.concat([df1, df2], ignore_index=True)
print(vertical_index)

# ==================================================
# 3. Horizontal Concatenation
# ==================================================
print("\n3. Horizontal Concatenation (axis=1)")
horizontal = pd.concat([df1, df3], axis=1)
print(horizontal)

# ==================================================
# 4. Concatenate Multiple DataFrames
# ==================================================
print("\n4. Multiple DataFrames")
multiple = pd.concat([df1, df2, df3], ignore_index=True)
print(multiple)

# ==================================================
# 5. Outer Join (Default)
# ==================================================
print("\n5. Outer Join")
outer_join = pd.concat([df1, df3], join="outer", ignore_index=True)
print(outer_join)

# ==================================================
# 6. Inner Join
# ==================================================
print("\n6. Inner Join")
inner_join = pd.concat([df1, df3], join="inner", ignore_index=True)
print(inner_join)

# ==================================================
# 7. Keys (MultiIndex)
# ==================================================
print("\n7. Keys (MultiIndex)")
keys_concat = pd.concat([df1, df2], keys=["First", "Second"])
print(keys_concat)

# ==================================================
# 8. Verify Integrity
# ==================================================
print("\n8. Verify Integrity")

try:
    result = pd.concat([df1, df2], verify_integrity=True)
    print(result)
except ValueError as e:
    print("Error:", e)

# ==================================================
# 9. Series Concatenation
# ==================================================
print("\n9. Series Concatenation")

s1 = pd.Series([10, 20, 30], name="Series1")
s2 = pd.Series([40, 50, 60], name="Series2")

series_result = pd.concat([s1, s2], ignore_index=True)
print(series_result)

# ==================================================
# 10. Horizontal Series Concatenation
# ==================================================
print("\n10. Horizontal Series Concatenation")

series_horizontal = pd.concat([s1, s2], axis=1)
print(series_horizontal)

print("\n" + "=" * 60)
print("All concat() operations completed successfully!")
print("=" * 60)
```
