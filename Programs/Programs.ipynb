1. Find Average, Maximum, and Minimum Salary

```python
import pandas as pd

data = {
    "Employee": ["A", "B", "C", "D"],
    "Salary": [30000, 45000, 50000, 40000]
}

df = pd.DataFrame(data)

print("Average Salary:", df["Salary"].mean())
print("Maximum Salary:", df["Salary"].max())
print("Minimum Salary:", df["Salary"].min())
```

2. Filter Students with Marks Greater Than 80

```python
import pandas as pd

data = {
    "Name": ["Rahul", "Priya", "Amit", "Sneha"],
    "Marks": [75, 88, 92, 65]
}

df = pd.DataFrame(data)

result = df[df["Marks"] > 80]

print(result)
```
