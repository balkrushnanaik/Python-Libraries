1. Find Average, Maximum, and Minimum Salary

```python
import pandas as pd

data = {
    "Employee": ["A", "B", "C", "D"],
    "Salary": [30000, 45000, 50000, 40000]
}

df = pd.DataFrame(data)

print("Average Salary:", df["Salary"].mean())
print("Maximum Salary:", df["Salary"].max())
print("Minimum Salary:", df["Salary"].min())
```

2. Filter Students with Marks Greater Than 80

```python
import pandas as pd

data = {
    "Name": ["Rahul", "Priya", "Amit", "Sneha"],
    "Marks": [75, 88, 92, 65]
}

df = pd.DataFrame(data)

result = df[df["Marks"] > 80]

print(result)
```

3. Count Missing Values

```python
import pandas as pd
import numpy as np

data = {
    "Name": ["A", "B", "C", "D"],
    "Age": [20, np.nan, 22, np.nan]
}

df = pd.DataFrame(data)

print("Missing Values:")
print(df.isnull().sum())
```

4. Group By Department and Calculate Average Salary

```python
import pandas as pd

data = {
    "Department": ["IT", "HR", "IT", "Sales", "HR"],
    "Salary": [50000, 40000, 60000, 45000, 42000]
}

df = pd.DataFrame(data)

result = df.groupby("Department")["Salary"].mean()

print(result)
```

5. Create a Simple Sales Analysis Report

```python
import pandas as pd

data = {
    "Month": ["Jan", "Feb", "Mar", "Apr"],
    "Sales": [10000, 12000, 15000, 13000]
}

df = pd.DataFrame(data)

print("Total Sales:", df["Sales"].sum())
print("Average Sales:", df["Sales"].mean())
print("Highest Sales:", df["Sales"].max())
```
