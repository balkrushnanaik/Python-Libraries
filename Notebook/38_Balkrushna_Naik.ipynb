#### Q3. Data Visualization-Create a histogram for each feature in the dataset to illustrate the feature distributions Plot each histogram



```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,8))
# plt.subplots_adjust(hspace=0.5, wspace=0.4)



column1 = df[['Sepal_Length']]
plt.figure(figsize = (6,4))
plt.subplot(2,2,1) # rows, column, position
plt.hist(column1)
plt.title('Distribution of First Column')


# Column2 Sepal_Width

column2 = df[['Sepal_Width']]
plt.subplot(2,2,2) # rows, column, position
plt.hist(column2)
plt.title('Distribution of second Column')


# Column3 Petal Length

column3 = df[['Petal_Length']]
plt.subplot(2,2,3) # rows, column, position
plt.hist(column3)
plt.title("Distribution of Third Column")
# plt.show()

# Column4 Petal Width

column4 = df[['Petal_Width']]
plt.subplot(2,2,4) 
plt.hist(column4)
plt.title('Distribution of Fourth Column')


plt.tight_layout() # It automatically adjusts spacing between charts.
plt.show()


```
