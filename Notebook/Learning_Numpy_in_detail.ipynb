# Learning Numpy In Detailed

NumPy is a core Python library for numerical computing, built for handling large arrays and matrices efficiently. It is significantly faster than Python's built-in lists because it uses optimized C language style storage where actual values are stored at contiguous locations (not object reference).


```python
import numpy as np

list1 = [1,2,3,4,5,6]
print(type(list1))
```

```python
arr_list = np.array(list1)
print(arr_list)
```

```python
print(arr_list.ndim)
```

# 2. Two Dimensional Array Creation
   

```python
list1 = [1,2,3]
list2 = [3,4,5]

arr_list2 = np.array([list1,list2])
print(arr_list2)
```

```python
print(f"It is {(arr_list2.ndim)} dimensional array")
```

3. Three Dimensional Array
   

```python
list1 = [1,2,3]
list2 = [3,4,5]
list3 = [6,7,8]

arr_list3 = np.array([[list1, list2, list3]])
print(arr_list3)
```

```python
print(f"It is {(arr_list3.ndim)} dimensional array")
```

# 4. Array Indexing and Slicing

1. Array indexing is the same as accessing an array element.
2. You can access an array element by referring to its index number.
3. The indexes in NumPy arrays start with 0, meaning that the first element has index 0, and the second has index 1 etc.

```python
arr_in = np.array([1,2,3,4,5,6])
print(arr_in)

#Sum of array indexing
sum_in = arr_in[0] + arr_in[1]
print(sum_in)
```

4.2 Array Slicing

Slicing in python means taking elements from one given index to another given index.

We pass slice instead of index like this: [start:end].

We can also define the step, like this: [start:end:step].

If we don't pass start its considered 0

If we don't pass end its considered length of array in that dimension

If we don't pass step its considered 1

```python
arr_sl = np.array([1,2,3,4,5,6])
print(arr_sl)

print(arr_sl[1::2])

# (Start: end: step)
```
