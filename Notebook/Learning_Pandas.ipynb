# Python Pandas for Data Analyst


#### 1. What is pandas?
- Pandas is a Python library used for working with data sets.
- It has functions for analyzing, cleaning, exploring, and manipulating data.
- The name "Pandas" has a reference to both "Panel Data", and "Python Data Analysis" and was created by Wes McKinney in 2008.

#### 2. Why Use Pandas?
- Pandas allows us to analyze big data and make conclusions based on statistical theories.
- Pandas can clean messy data sets, and make them readable and relevant.
- Relevant data is very important in data science.

#### 3. What Can Pandas Do?
- Pandas gives you answers about the data. Like:
- Is there a correlation between two or more columns?
- What is average value?
- Max value?
- Min value?
Pandas are also able to delete rows that are not relevant, or contains wrong values, like empty or NULL values. This is called cleaning the data.

# 4 Rows & Columns - Selection

```python
import pandas as pd 

data = {
    'Name': ['Shreya', 'Priya', 'Jiya', 'Riya', 'Ram', 'Karan'],
    'Age': [22, 23, 20, 18, 19, 24],
    'Gender': ['F', 'F', 'F', 'F', 'M', 'M']
}

df = pd.DataFrame(data)
print(df)
```

```python
# Select column

# df[['Name']] # For selecting single column
df[['Name', 'Age']] # For selecting two columns
```

```python
# select single row || loc - label based index
df.loc[df.Name=='Priya']
```

```python
# select multiple rows || loc - label based index

df.loc[0:4]
```

```python
# Select Rows || iloc - index-value based
df.iloc[0]
# df.iloc[0:2] # [start:stop(less than 1):step]
```

# 5. Filter Dataframe - Filtering by column values

```python
df
```

```python
df[(df['Age']>=20) & (df['Gender']== 'F')]
```

#### Where(): 
- The where() function in Pandas is used to keep values that satisfy a specified condition and replace values that do not satisfy the condition with another value.

```python


df.where(df['Age']>20, other='less than 20')
```

# 6 Rows and Columns - Operation (Add, update, delete)

```python
import pandas as pd
data = {
    "Name": ['Soumya','Vaibhav','Himanshu','Pranit','Shivani','Rakesh'],
    "Age": [22,23,24,23,22,25],
    "salary": [22000,45000,67000,100000,150000,50000]

}
df = pd.DataFrame(data)
print(df)
```

```python
# Create a new column
df['Team']= ['Analyst','Developer','Testing','Sales','Cloud','AI']
df
```

```python
# Add new columns using existing column
df['Bonus'] = df['salary'] * 0.30
df
```

```python
df.rename(columns= {'salary':'Salary'}, inplace = True)
```

```python
df
```

```python
# Add new row - at the end of dataframe
df.loc[len(df)] = ['Janki', 21, 21000, 'IT', 2000]
df
```

```python
# Update value in Dataframe using column value
df.loc[df.Age==25,'Salary']=75000
df
```

```python
df.loc[df.Team == 'IT', 'Salary'] = 80000
df
```

#### Delete value - rows and columns

#### drop():
- The drop() function is used to delete unwanted rows or columns from a Pandas DataFrame based on their labels or indexes. It is commonly used during data cleaning and preprocessing.

```python
# delete row using column-value filter
df = df.drop(df[df.Name == 'Janki'].index)
print(df)
```

```python
# delete row using index-name filter
df.drop(5, axis=0)
```

#### sort_values(): 
- The sort_values() function is used to arrange DataFrame rows based on the values of one or more columns in ascending or descending order. It helps organize and analyze data more effectively.

```python
df.sort_values('Salary') # Ascending order, by default
```

```python
df.sort_values(['Age','Salary'],ascending = False)
```
