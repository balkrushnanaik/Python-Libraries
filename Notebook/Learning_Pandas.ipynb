# 7. Pandas - Cleaning Data


- Data cleaning means fixing bad data in your data set.
- Bad data could be:

Empty cells,
Data in wrong format,
Wrong data,
Duplicates etc.



#### 7.1 Pandas - Cleaning Empty Cells


- Empty cells can potentially give you a wrong result when you analyze data.

```python
df1 = pd.read_csv('data_set.csv')
print(df1.to_string())
```

```python
new_df = df1.dropna()
print(new_df.to_string())

# By default, the dropna() method returns a new DataFrame, and will not change the original.
# If you want to change the original DataFrame, use the inplace = True argument:

```

Note: Now, the dropna(inplace = True) will NOT return a new DataFrame, but it will remove all rows containing NULL values from the original DataFrame.

##### Replace Empty Values


- Another way of dealing with empty cells is to insert a new value instead.
- This way you do not have to delete entire rows just because of some empty cells.
- The fillna() method allows us to replace empty cells with a value:





```python
print(df1.isnull().sum())
```

```python
df = pd.read_csv('data_set.csv')
print(df.to_string())
```

```python
# Display missing values
print(df.isnull().sum())

# Fill missing values

# Fill missing names with "Unknown"
df["name"] = df["name"].fillna("Unknown")

# Fill missing age with average age
df["age"] = pd.to_numeric(df["age"], errors="coerce")
df["age"] = df["age"].fillna(df["age"].mean())

# Fill missing salary with average salary
df["salary"] = pd.to_numeric(df["salary"], errors="coerce")
df["salary"] = df["salary"].fillna(df["salary"].mean())

# Fill missing joining_date with "Not Available"
df["joining_date"] = df["joining_date"].fillna("Not Available")

# Fill missing department with "General"
df["department"] = df["department"].fillna("General")

# Display updated dataset
print(df)
```
