Learning Numpy In Detailed

```python
import numpy as np

list1 = [1,2,3,4,5,6]
print(type(list1))
```

```python
arr_list = np.array(list1)
print(arr_list)
```

```python
print(arr_list.ndim)
```

.ndim is used check how many dimensional array is this

2. Two Dimensional Array Creation
   

```python
list1 = [1,2,3]
list2 = [3,4,5]

arr_list2 = np.array([list1,list2])
print(arr_list2)
```

```python
print(f"It is {(arr_list2.ndim)} dimensional array")
```

3. Three Dimensional Array
   

```python
list1 = [1,2,3]
list2 = [3,4,5]
list3 = [6,7,8]

arr_list3 = np.array([[list1, list2, list3]])
print(arr_list3)
```

```python
print(f"It is {(arr_list3.ndim)} dimensional array")
```

2. Numpy Array Attributes

```python
import numpy as np

arr1 = np.array([[[1.1,2.2,3.1], [4.1,5.1,6.1], [7.1,7.4,7.8]]])
print(type(arr1))
```

```python
print(f"Shape {arr1.shape}")
print(arr1)
print(f"It is {arr1.ndim} dimensional array")
print(f"size {arr1.size}")
```
