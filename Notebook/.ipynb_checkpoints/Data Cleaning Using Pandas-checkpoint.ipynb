```python
import pandas as pd
import numpy as np

data = pd.read_csv('sample.csv')
print(data.to_string())
```

```python
data.fillna({'Name':'Unknown'}, inplace=True)
print(data.to_string())
```

```python
data["Age"] = pd.to_numeric(data["Age"], errors="coerce")
data["Age"] = data["Age"].fillna(data["Age"].mean())
```

```python
print(data.to_string())
```

```python
data.fillna({'Email': 'unknown00@gmail.com'}, inplace= True)
print(data.to_string())
```

```python
def clean_email(email):
    if isinstance(email, str) and "@" in email and "." in email:
        return email
    else:
        return "unknown@gmail.com"

data["Email"] = data["Email"].apply(clean_email)
```

```python
print(data.to_string())
```

```python

data["Salary"] = pd.to_numeric(data["Salary"], errors="coerce")

avg_salary = data["Salary"].mean()

def clean_salary(salary):
    if salary != 'Invalid':
        return salary
    else:
        return avg_salary

data["Salary"] = data["Salary"].apply(clean_salary)
print(data.to_string())
```

```python
data['City'] = data['City'].fillna('Unknown')
print(data.to_string())
```

```python
data['Join_Date'] = pd.to_datetime(data['Join_Date'], errors='coerce')
data['Join_Date'] = data['Join_Date'].fillna('2024-01-01')
print(data.to_string())
```

# Today I have learned how to clean data in Pandas
