# Seaborn Library in Python
#### Introduction

- Seaborn is a powerful Python library used for data visualization. It is built on top of Matplotlib and provides a high-level interface for creating attractive and informative statistical graphics.
- Seaborn works very well with Pandas DataFrames.

##### Importing Library

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

##### Loading Dataset

```python
import seaborn as sns

data = sns.load_dataset("tips")

print(data.head())
```

##### Important Seaborn Plots

```python
sns.boxplot( x = 'sex', y = 'total_bill', data = data)
plt.show()
```
